```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("BAAI/bge-m3",device = "cpu")
```

```python
import pandas as pd
texts = pd.read_csv('train.csv')['question'].values.tolist()
```

```python
texts[0]
```

```python
pool = model.start_multi_process_pool(["cuda:0","cuda:1"])
```

```python
import torch
torch.cuda.empty_cache()
```

```python

emb = model.encode_multi_process(texts,pool,batch_size = 8, show_progress_bar = True)
print("Embeddings computed. Shape:", emb.shape)

```

```python
model.stop_multi_process_pool(pool)
```

```python
import numpy as np
np.save('train_emb.npy', emb)
```
